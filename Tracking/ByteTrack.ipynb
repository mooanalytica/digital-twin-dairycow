```python
from ultralytics import YOLO
import os

# Paths
model_path = "best.pt"
video_dir = "clips4"
save_dir = "track_results4"

# Load your trained YOLO model
model = YOLO(model_path)
```

```python
# ByteTrack config path (included with ultralytics)
bytetrack_yaml = "ultralytics/cfg/trackers/bytetrack.yaml"
```

```python
# Create output directory
os.makedirs(save_dir, exist_ok=True)

# Loop through all 10s clips
for filename in os.listdir(video_dir):
    if filename.endswith(".mp4"):
        video_path = os.path.join(video_dir, filename)
        print(f"Processing: {video_path}")

        # Run YOLO + ByteTrack
        model.track(
            source=video_path,
            tracker=bytetrack_yaml,
            save=True,
            save_txt=True,
            save_conf=True,
            imgsz=640,
            conf=0.3,
            project=save_dir,
            name=filename.split(".")[0]
        )
```
