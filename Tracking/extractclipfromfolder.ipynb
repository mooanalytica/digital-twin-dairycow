```python
import os
import shutil

# Source directory (replace with your actual path)
source_dir = 'PaddedCowVideos4'  # This is the root folder containing nested folders
# Destination directory
destination_dir = 'shreyaannotate'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# Traverse the directory tree
for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file.endswith('.mp4'):
            source_path = os.path.join(root, file)
            # Ensure unique filename to avoid overwriting
            unique_name = os.path.relpath(source_path, source_dir).replace(os.sep, '_')
            dest_path = os.path.join(destination_dir, unique_name)
            shutil.copy2(source_path, dest_path)
            print(f"Copied: {source_path} → {dest_path}")
```
